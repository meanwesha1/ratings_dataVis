```python
#We will work on a dataset containing movie ratings from different websites.
#We will analyze the ratings given by different websites and check whether there is any kind of bias or not.
```

```python
import pandas as pd
reviews = pd.read_csv('fandango_scores.csv')
pd.options.display.max_columns = 99
pd.options.display.width = 100
print(reviews.head())
```

```python
#We will select few of the columns for our purpose.
```

```python
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
norm_reviews = reviews[cols]
print(norm_reviews.head())
```

```python
#According to sources, the star ratings in Fandango are expressed at half-star increments from the average rating value.
#To check that, let's compare the ratings for 5 random movies.
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,6))
sampledf = norm_reviews.sample(5)
from numpy import arange
bar_positions = arange(5) + 0.95
bar_heights = sampledf['Fandango_Ratingvalue'].values
bar_width = 0.25
ax.bar(bar_positions, bar_heights, bar_width, label='Fandango_Ratingvalue')
bar_heights = sampledf['Fandango_Stars'].values
ax.bar(bar_positions+bar_width, bar_heights, bar_width, label='Fandango_Stars')
tick_positions = range(1,6)
ax.set_xticks(tick_positions)
tick_positions = arange(0,6,0.5)
ax.set_yticks(tick_positions)
ax.set_xticklabels(sampledf['FILM'], rotation=5)
ax.set_xlabel('Films')
ax.set_ylabel('Ratings')
ax.set_title('Star Rating and Average User Rating For 5 sample movies.')
ax.legend()
plt.show()
```

```python
#Well it seems like the news is true, the number of stars given to the movies is rounded up by .5 above its numerical rating.
#Now since Fandango sells movie tickets directly to customers,
#it might have higher interest in making people intrigued at movies with their good ratings even if the movies are not that good.
#Let's check the ratings given by the 5 websites to the first movie.
```

```python
fig, ax = plt.subplots(figsize=(10,6))
bar_heights = norm_reviews.iloc[0,1:].values
bar_positions = arange(5) + 0.75
ax.bar(bar_positions, bar_heights, 0.5)
tick_positions = range(1,6)
ax.set_xticks(tick_positions)
ax.set_xticklabels(cols[1:], rotation=20)
ax.set_xlabel('Rating Source')
ax.set_ylabel('Average Rating')
ax.set_title('Average User Rating For Avengers: Age of Ultron (2015)')
plt.show()
```

```python
#We can see that RT_user_norm and Fandango_Ratingvalue has given higher ratings compared to the other 2 websites.
#We would like to check whether the same happens for all the movies or not.
#Also, we would like to see how Metacritic and IMDB agree with Fandango on movie-ratings on average.
```

```python
fig = plt.figure(figsize = (12,6))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)
ax1.scatter(norm_reviews['Fandango_Ratingvalue'],norm_reviews['RT_user_norm'])
ax2.scatter(norm_reviews['Fandango_Ratingvalue'],norm_reviews['Metacritic_user_nom'])
ax3.scatter(norm_reviews['Fandango_Ratingvalue'],norm_reviews['IMDB_norm'])
ax1.set_xlabel('Fandango_Ratingvalue')
ax1.set_ylabel('RT_user_norm')
ax2.set_xlabel('Fandango_Ratingvalue')
ax2.set_ylabel('Metacritic_user_nom')
ax3.set_xlabel('Fandango_Ratingvalue')
ax3.set_ylabel('IMDB_norm')
ax1.set_xlim(0,5)
ax2.set_xlim(0,5)
ax3.set_xlim(0,5)
ax1.set_ylim(0,5)
ax2.set_ylim(0,5)
ax3.set_ylim(0,5)
plt.show()
```

```python
#We can clearly see that most of the user ratings in Fandango are within the range 3 and 5,
#while ratings in RottenTomatoes, Metacritic and IMDB span from around 1 or 2 to 5.
#This shows the skewness in Fandango ratings.
#It gives us a good example of why one should be skeptical of online movie ratings, especially from companies selling tickets.
```
